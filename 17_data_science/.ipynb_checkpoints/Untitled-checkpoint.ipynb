```python
import pandas as pd
```

```python
contoso_cadastro_produtos = pd.read_csv(r'Contoso - Cadastro Produtos.csv', sep=';')
contoso_clientes = pd.read_csv(r'Contoso - Clientes.csv', sep=';')
contoso_lojas = pd.read_csv(r'Contoso - Lojas.csv', sep=';')
contoso_promocoes = pd.read_csv(r'Contoso - Promocoes.csv', sep=';')
contoso_vendas_2017 = pd.read_csv(r'Contoso - Vendas - 2017.csv', sep=';')
```

```python
# first step -> treat the data

display(contoso_vendas_2017)  # main dataframe 
display(contoso_cadastro_produtos)
display(contoso_clientes)
display(contoso_lojas)
display(contoso_promocoes)
```

```python
contoso_cadastro_produtos = contoso_cadastro_produtos[['Nome da Marca', 'ID Produto']]
contoso_clientes = contoso_clientes[['Primeiro Nome', 'E-mail', 'ID Cliente']]
contoso_lojas = contoso_lojas[['Nome da Loja', 'ID Loja']]
contoso_promocoes = contoso_promocoes[['Nome Promocao', 'ID Promocao']]

display(contoso_cadastro_produtos)
display(contoso_clientes)
display(contoso_lojas)
display(contoso_promocoes)
display(contoso_vendas_2017)
```

```python
contoso_vendas_2017 = contoso_vendas_2017.merge('contoso_cadastro_produtos', on='ID Produto')
contoso_vendas_2017 = contoso_vendas_2017.merge('contoso_clientes', on='ID Cliente')
contoso_vendas_2017 = contoso_vendas_2017.merge('contoso_lojas', on='ID Loja')
contoso_vendas_2017 = contoso_vendas_2017.merge('contoso_promocoes', on='ID Promocao')
```
